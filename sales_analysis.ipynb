# **Converting Date Columns**


```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
```

```python
data.info()
```

```python
data ["Order Month"] = data["Order Date"].dt.month
data ["Order year"] = data ["Order Date"].dt.year
data ["Order day of week"] = data["Order Date"].dt.dayofweek
```

```python
data.head()
```

#**Monthly Sales Analysis**


```python
sales_by_month = data.groupby("Order Month")["Sales"].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month, x="Order Month", y="Sales", title="Monthly Sales Analysis")
fig.show()
```

#**Sales Analysis by Category**


```python
sales_by_category = data.groupby("Category")["Sales"].sum().reset_index()
```

```python
sales_by_category
```

```python
fig = px.pie(sales_by_category,
             values="Sales",
             names="Category",
             title="Sales by Category",
             color_discrete_sequence= px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_font=dict(size=24))
fig.show()
```

#**Sales Analysis by Sub Category**


```python
sales_by_subcategory = data.groupby("Sub-Category")["Sales"].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x= "Sub-Category", y = "Sales", title= "Sales Analysis by Sub-Category")
fig.show()
```

#**Sales Analysis by Category and Sub Category**


```python
sales_by_subcategory_and_category = (
    data.groupby(["Category", "Sub-Category"])["Sales"]
    .sum()
    .reset_index()
)
```

```python
sales_by_subcategory_and_category
```

```python
import plotly.express as px

fig = px.bar(
    sales_by_subcategory_and_category,
    x="Sub-Category",
    y="Sales",
    color="Category",
    barmode="group",  # or "stack"
    title="Sales by Sub-Category and Category"
)

fig.show()
```

#**Sales and Profit - Customer Segment**

```python
sales_profit_by_segment = (
    data.groupby("Segment")
    .agg({"Sales": "sum", "Profit": "sum"})
    .reset_index()
)
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=sales_profit_by_segment["Segment"],
        y=sales_profit_by_segment["Sales"],
        name="Sales",
        marker_color=color_palette[0],
    )
)
fig.add_trace(
    go.Bar(
        x=sales_profit_by_segment["Segment"],
        y=sales_profit_by_segment["Profit"],
        name="Profit",
        marker_color=color_palette[1],
    )
)
fig.update_layout(
    title="Sales and Profit by Customer Segment",
    xaxis_title="Customer Segment",
    yaxis_title="Amount",
    barmode="group",
)
fig.show()
```

#**Sales to profit ratio**

```python
sales_profit_by_segment = (
    data.groupby("Segment")
    .agg({"Sales": "sum", "Profit": "sum"})
    .reset_index()
)

sales_profit_by_segment["Sales to Profit Ratio"] = (
    sales_profit_by_segment["Sales"] / sales_profit_by_segment["Profit"]
)
print(sales_profit_by_segment[['Segment', 'Sales to Profit Ratio']])
```
